# Locate card 

Locate card in a list of cards ordered in descend order. Binary search.

```python
def locate_card_linear_search(list_of_cards, card):

    len_list = len(list_of_cards) 
    
    position = 0
    result = -1
    
    while(0 <= position < len(list_of_cards)):

        if len_list == 0 or position >= len_list:
            result = -1
            break

        if(list_of_cards[position] == card):
            result = position
            break
        
        position += 1
    
    return result
```

```python
def locate_card_recursive(list_of_cards, card):
    len_list = len(list_of_cards)
#     print(myList)
    mid = len_list//2
#     print(mid)
    solution = -1
    if len_list == 0:
        solution = -1
    elif mid == 0:
        if list_of_cards[mid] == card:
            solution = mid
    elif mid > 0 and list_of_cards[mid - 1] == card:
        new_list = list_of_cards[:mid]
        solution = locate_card(new_list, card)
    elif list_of_cards[mid] == card:
        solution = mid
    elif list_of_cards[mid] < card:
        new_list = list_of_cards[:mid]
        solution = locate_card(new_list, card)
    else:
        new_list = list_of_cards[mid:]
        solution = mid + locate_card(new_list, card)
#     print(solution)
    return solution
```

```python
def locate_card_binary(list_of_cards, card):
        len_list = len(list_of_cards)
        
        if len_list == 0:
            return -1
        
        low_index = 0
        high_index = len_list - 1
        result = -1
        
        while low_index <= high_index:
            mid = (high_index + low_index) // 2
            
            if mid > 0 and list_of_cards[mid - 1] == card:
                high_index = mid - 1
            elif list_of_cards[mid] == card:
                result = mid
                break
            elif list_of_cards[mid] < card:
                high_index = mid - 1
            elif list_of_cards[mid] > card:
                low_index = mid + 1
        return result
```
